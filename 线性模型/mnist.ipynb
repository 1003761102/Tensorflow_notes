```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from  tensorflow.examples.tutorials.mnist import input_data
```

```python
mnist = input_data.read_data_sets('MNIST/', one_hot=True)
```

```python
print(mnist.train.num_examples)
print(mnist.test.num_examples)
```

```python
train_image = mnist.train.images
train_label = mnist.train.labels
test_image = mnist.test.images
test_label = mnist.test.labels
print(train_image.shape)
print(train_label.shape)
```

```python
nsample = 5
randinx = np.random.randint(train_image.shape[0], size=nsample)
print(randinx)
for inx in randinx:
    image = np.reshape(train_image[inx,:], (28, 28))
    label = np.argmax(train_label[inx, :])
    plt.matshow(image, cmap=plt.get_cmap('gray'))
    plt.title('The number '+str(inx)+' image label is:'+str(label))
    plt.show()
```
